# USING FUNCTION TO ANALYSE UTIVA BREWERIES DATA

```python
import time as time
import pandas as pd
import numpy as np

# profit worth of the breweries for the last 3 years
def load_data():
    print('hello let\'s explore some utiva breweries data')
    info=input('enter a data file:')
    return info

def low_data(info):
    df=pd.read_csv(info)
    return df

def filters(df):
    sumprofit=df['PROFIT'].sum()
    print(sumprofit)
    print('-'*40)
    
#     compare profits for francophone and anglopgone countries
def franc(df):
    df2=df[(df.COUNTRIES=='Benin') | (df.COUNTRIES=='Senegal') | (df.COUNTRIES=='Togo')]
    df_anglo=df[(df.COUNTRIES=='Nigeria') | (df.COUNTRIES=='Ghana')]
    df2profit=df2['PROFIT'].sum()
    dfang=df_anglo['PROFIT'].sum()
    print(f'the profit for anglophone is {df2profit}')
    print(f'the profit for francophone is {dfang}')
    print('-'*40)
    
    #Country that generated the highest profit in 2019

def prof_country(df):
    df_19=df[(df.YEARS==2019)]
    df_pro=df_19.groupby('COUNTRIES')['PROFIT'].sum().sort_values(ascending=False)
    dff=df_pro.head(1)
    print(f'the country with the highest profit for 2019 is {dff}')
    print('-'*40)
    
#      find the year with the highest profit
def prof_year(df):
    df_19=df[(df.YEARS==2019)]
    dff_19=df_19['PROFIT'].sum()
    df_18=df[(df.YEARS==2018)]
    dff_18=df_18['PROFIT'].sum()
    df_17=df[(df.YEARS==2017)]
    dff_17=df_17['PROFIT'].sum()
    if dff_19 > dff_18 and dff_19 > dff_17:
        print(f' the year with the highest profit is 2019 {dff_19}')
    elif dff_18 > dff_19 and dff_18 > dff_17:
        print(f' the year with the highest profit is 2018 {dff_18}')
    elif dff_17 > dff_19 and dff_17 > dff_18:
        print(f'the year with the highest profit is 2017 {dff_17}')
    print('-'*40)
    
# Which month in the three years was the least profit generated?

def mont(df):
    mon=df.groupby('MONTHS')['PROFIT'].sum().sort_values(ascending=False)
    df_mon=mon.reset_index()
    x=df_mon['PROFIT'].min()
    p=df_mon[(df_mon.PROFIT==x)]
    print(f'month with the least profit')
    print(p)
    print('_'*40)

def perc(df):
    month_b = df[(df.YEARS==2019)]
    df5=month_b
    sum_profit = df5['PROFIT'].sum()
    arr_profit=np.array(df5['PROFIT'])
    for i in arr_profit:
        percent = ((i/sum_profit)*100)
    print(f'the profit for {i} is : {round(percent,2)}%')
    print('_'*40)

#     minimum profit for december 2018
def dec(df):
    df_dec=df[(df.MONTHS=='December')]
    dff_dec=df_dec[(df_dec.YEARS==2018)]
    dff_min=dff_dec['PROFIT'].min()
    print(f' the minimum profit for december 2018 is {dff_min}')
    print('_'*40)

# Which particular brand generated the highest profit in Senegal?
def senegal(df):
    df_sen=df[(df.COUNTRIES=='Senegal')]
    dff_sen=df_sen.groupby('BRANDS')['PROFIT'].sum().sort_values(ascending=False).head(1)
    print(f'the brand with the highest profit in senegal is {dff_sen}')
    print('_'*40)
    
# BRAND ANALYSIS
# Within the last two years, the brand manager wants to know the top three brands
# consumed in the francophone countries

def brand(df):
    df_brand=df[(df.YEARS==2017)|(df.YEARS==2018)]
    df_fran=df_brand[(df_brand.COUNTRIES=='Benin')|(df_brand.COUNTRIES=='Senegal')|(df_brand.COUNTRIES=='Togo')]
    dff_fran=df_fran.groupby('BRANDS')['PROFIT'].sum().sort_values(ascending=False).head(3)
    print(f'the top three brands in the francophone countries is {dff_fran}')
    print('_'*40)
    
# Find out the top two choice of consumer brands in Ghana

def gh(df):
    df_ghana=df[(df.COUNTRIES=='Ghana')]
    dff_ghana=df_ghana.groupby('BRANDS')['QUANTITY'].sum().sort_values(ascending=False).head(2)
    print(f'the top two choice of consumer brands in ghana is {dff_ghana}')
    print('_'*40)
    
# Find out the details of beers consumed in the past three years in the most oil reached
# country in West Africa.
def details(df):
    df_nig=df[(df.COUNTRIES=='Nigeria')]
    df_beer=df_nig[(df_nig.BRANDS=='trophy')|(df_nig.BRANDS=='budweiser')|(df_nig.BRANDS=='castle lite')
                   |(df_nig.BRANDS=='eagle lager')|(df_nig.BRANDS=='hero')]
    dff_beer=df_beer.head(4)
    print(df_beer)
    print('_'*40)
    
# Favorites malt brand in Anglophone region between 2018 and 2019
def fav_malt(df):
    df_malt=df[(df.YEARS==2019)|(df.YEARS==2018)]
    dff_malt=df_malt[(df_malt.COUNTRIES=='Nigeria')|(df_malt.COUNTRIES=='Ghana')]
    df_fav=dff_malt[(dff_malt.BRANDS=='beta malt')|(dff_malt.BRANDS=='grand malt')]
    dff_fav=df_fav.groupby('BRANDS')['QUANTITY'].sum().sort_values(ascending=False).head(1)
    print(f' the favourite malt brand in anglophone between 2018 and 2019 is {dff_fav}')
    print('_'*40)
    
# Which brands sold the highest in 2019 in Nigeria?
def high_brand(df):
    df_high=df[(df.YEARS==2019)]
    dff_high=df_high[(df_high.COUNTRIES=='Nigeria')]
    dff_hig=dff_high.groupby('BRANDS')['QUANTITY'].sum().sort_values(ascending=False).head(4)
    print(f' the brands that sold the highest in nigeria in 2019 are {dff_hig}')
    print('_'*40)
    
# Favorites brand in South_South region in Nigeria
def south(df):
    ng=df[(df.COUNTRIES=='Nigeria')]
    dng=ng["REGION "] =="southsouth"
    fff=ng[dng]
    fav_ng=fff.groupby('BRANDS')['QUANTITY'].sum().sort_values(ascending=False).head(1)
    print(f'the favourite brand in south south in nigeria is {fav_ng}')
    print('_'*40)
    
# Bear consumption in Nigeria
def beer_nig(df):
    bng=df[(df.COUNTRIES=='Nigeria')]
    df_beer=bng[(bng.BRANDS=='trophy')|(bng.BRANDS=='budweiser')|(bng.BRANDS=='castle lite')
                   |(bng.BRANDS=='eagle lager')|(bng.BRANDS=='hero')]
    print(f' details of beer consumption in nigeria \n{df_beer}')
    print('_'*40)
    
# Level of consumption of Budweiser in the regions in Nigeria
def level(df):
    dnig=df[(df.COUNTRIES=='Nigeria')]
    dfn=dnig[(dnig.BRANDS=='budweiser')]
    dffn=dfn.groupby('BRANDS')['QUANTITY'].sum()
    print(f'the level of budweiser consumption in the regions in nigeria \n {dffn}')
    print('_'*40)
    
    
# Level of consumption of Budweiser in the regions in Nigeria in 2019 (Decision on Promo
def bud(df):
    dnig=df[(df.COUNTRIES=='Nigeria')]
    dfn=dnig[(dnig.BRANDS=='budweiser')]
    ddf=dfn[dfn.YEARS==2019]
    print(f' level of budweiserr consumption in regions in nigeria in 2019 \n {ddf}')
    print('_'*40)
    
# Country with the highest consumption of beer.
def con(df):
    df_con=df[(df.BRANDS=='trophy')|(df.BRANDS=='budweiser')|(df.BRANDS=='castle lite')
                   |(df.BRANDS=='eagle lager')|(df.BRANDS=='hero')]
    dff_con=df_con.groupby('COUNTRIES')['QUANTITY'].sum().sort_values(ascending=False).head(1)
    print(f' the country with the highest consumption of beer is {dff_con}')
    print('_'*40)

# Highest sales personnel of Budweiser in Senegal
def sales(df):
    dfsales=df[(df.COUNTRIES=='Senegal')]
    dff_sales=dfsales[(dfsales.BRANDS=='budweiser')]
    dff_s=dff_sales.groupby('SALES_REP')['QUANTITY'].sum().sort_values(ascending=False).head(1)
    print(f' the highest sales personnel of budweiser in senegal is {dff_s}')
    print('_'*40)
    
# Country with the highest profit of the fourth quarter in 2019
def quarter(df):
    dfq=df[df.YEARS==2019]
    dfquat=dfq[(dfq.MONTHS=='September')|(dfq.MONTHS=='October')|(dfq.MONTHS=='November')|(dfq.MONTHS=='December')]
    dfc=dfquat.groupby('COUNTRIES')['PROFIT'].sum().sort_values(ascending=False).head(1)
    print(f'the country with the highest profit for 2019 is {dfc}')
    print('_'*40)





def main():
    func=load_data()
    defunct=low_data(func)   
    fun1=filters(defunct)
    fun2=franc(defunct)
    fun3=prof_country(defunct)
    fun4=prof_year(defunct)
    fun5=mont(defunct)
    fun6=perc(defunct)
    fun7=dec(defunct)
    fun8=senegal(defunct)
    fun9=brand(defunct)
    fun10=gh(defunct)
    fun11=details(defunct)
    fun12=fav_malt(defunct)
    fun13=high_brand(defunct)
    fun14=south(defunct)
    fun15=beer_nig(defunct)
    fun16=level(defunct)
    fun17=bud(defunct)
    fun18=con(defunct)
    fun19=sales(defunct)
    fun20=quarter(defunct)
    
if __name__ == "__main__":
    main()
```
